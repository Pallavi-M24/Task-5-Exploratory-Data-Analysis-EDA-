```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Survived"].value_counts()
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"].dropna(), bins=20, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Fare"])
plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.title("Age vs Fare")
plt.show()
```

```python
df_encoded = df.copy()
df_encoded["Sex"] = df_encoded["Sex"].map({"male": 0, "female": 1})

plt.figure(figsize=(10,6))
sns.heatmap(df_encoded.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[["Age", "Fare", "Pclass", "Survived"]])
plt.show()
```

```python
## Titanic EDA Summary

- The dataset contains passenger details like age, sex, class, fare, and survival.
- Many missing values exist in Age and Cabin columns.
- Females had a higher survival rate compared to males.
- First-class passengers had better chances of survival.
- Fare shows outliers indicating premium tickets.
- Passenger class and gender strongly influenced survival.
```
